```python
def inicio():
  print("*Inicia el duelo*")

def giro_vertical(n):
  for i in range (0,n):
    print("*Giro vertical*")

def giro_enclenque(n):
  for i in range (0,n):
    print("*Giro enclenque*")

def ataque(n):
  for i in range (0,n):
    print("*Ataque*")

def mantener(s):
  print("*Mantiene por",s,"segundos*")

def retroceder():
  print("*Retrocede*")

def caminar_circulos(s):
  print("*Camina en circulos por",s,"segundos*")

def final():
  print("*Termina el duelo*")

def giro_espalda(n):
  for i in range (0,n):
    print("*Giro espalda*")

def giro_espalda_enclenque(n):
  for i in range (0,n):
    print("*Giro espalda enclenque*")

def pose_de_batalla_epica():
  print("*Pose de batalla épica*")

def duelo1():
  print("Comienza el duelo 1")
  inicio()
  giro_vertical(2)
  giro_enclenque(2)
  ataque(4)
  mantener(3)
  retroceder()
  caminar_circulos(5)
  giro_vertical(5)
  giro_enclenque(4)
  ataque(4)
  mantener(3)
  retroceder()
  giro_vertical(3)
  giro_enclenque(2)
  caminar_circulos(7)
  giro_vertical(5)
  giro_vertical(5)
  ataque(4)
  mantener(4)
  retroceder()
  giro_vertical(3)
  giro_enclenque(3)
  caminar_circulos(7)
  giro_vertical(8)
  giro_enclenque(4)
  ataque(4)
  mantener(3)
  giro_vertical(2)
  giro_enclenque(2)
  final()

def duelo2():
  print("Comienza el duelo 2")
  inicio()
  giro_espalda(1)
  giro_espalda_enclenque(2)
  pose_de_batalla_epica()
  ataque(6)
  retroceder()
  giro_espalda(3)
  giro_espalda_enclenque(4)
  pose_de_batalla_epica()
  caminar_circulos(4)
  ataque(6)
  retroceder()
  giro_espalda(4)
  giro_espalda_enclenque(4)
  pose_de_batalla_epica()
  caminar_circulos(5)
  ataque(6)
  final()

def duelo_completo():
  for i in range(0,2):
    duelo1()
    duelo2()
```

```python
duelo_completo()
```
